```
url='https://www.linkedin.com/jobs/'
browser.get(url)

soup = BeautifulSoup(browser.page_source)
```

```
soup
```

```
job_title=soup.find_all('div',class_='flex-grow-1 artdeco-entity-lockup__content ember-view')
job_title
```

```
job_list=[]
for job in job_title:
    j=job.find('div',class_='job-card-square__title').text.replace('\n','').replace('Job Title','').replace('                                                  ','').replace('              ','')
    company_name=job.find('div',class_='job-card-container__company-name').text.replace('\n                ','').replace('\n              ','')
    location=job.find('li',class_='job-card-container__metadata-item').text
    job_list.append({"Job title":j,
                    "Comapny_name":company_name,
                     "Location":location
                    })
    
```
